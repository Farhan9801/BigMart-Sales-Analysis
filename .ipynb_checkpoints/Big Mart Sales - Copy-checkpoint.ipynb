```python
import numpy as np
from scipy import stats

def count_outliers(data, threshold=3):
    z_scores = np.abs(stats.zscore(data))
    outliers = np.where(z_scores > threshold)
    return len(outliers[0])



outlier_count = count_outliers(big_mart_data['Item_Outlet_Sales'])
print("Number of outliers:", outlier_count)
```

Handling Missing Values

Mean --> average value
Mode --> Most repeated value

```python
# mean value of 'Item_Weight' column

big_mart_data_without_outliers['Item_Weight'].mean()
```

```python
#  filling the missing value in the 'Item_Weight' column with "Mean value"

big_mart_data_without_outliers['Item_Weight'].fillna(big_mart_data_without_outliers['Item_Weight'].mean(), inplace = True)
```

```python
# checking for missing valueb

big_mart_data_without_outliers.isnull().sum()
```

```python
# mode of "Outlet_Size" column
big_mart_data_without_outliers['Outlet_Size'].mode()
```

Replacing the missing values in "Outlet_Size" with Mode

```python
# filling the missing values in "Outlet_Size" column with Mode
mode_of_Outlet_size = big_mart_data_without_outliers.pivot_table(values='Outlet_Size', columns='Outlet_Type', aggfunc=(lambda x: x.mode()[0]))
```

```python
print(mode_of_Outlet_size)
```

```python
miss_values = big_mart_data_without_outliers['Outlet_Size'].isnull()

print(miss_values)
```

```python
big_mart_data_without_outliers.loc[miss_values, 'Outlet_Size'] = big_mart_data_without_outliers.loc[miss_values,'Outlet_Type'].apply(lambda x: mode_of_Outlet_size[x])
```

Splitting Feature and Target

```python
X = big_mart_data_without_outliers.drop(columns = 'Item_Outlet_Sales', axis = 1)
Y = big_mart_data_without_outliers['Item_Outlet_Sales']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = )
```

```python
print(X.shape, x_train.shape, x_test.shape)
```
